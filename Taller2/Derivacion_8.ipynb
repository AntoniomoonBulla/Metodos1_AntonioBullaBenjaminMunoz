```python
import numpy as np
import matplotlib.pyplot as plt

h = 0.01
x = np.linspace(0.1,1.1,100)

def Function(x):
    return np.sqrt(np.tan(x))

def ProgressiveDerivative(f,x,h):
    
    d = 0.
    
    if h != 0:
        d = (-3*f(x)+4*f(x+h)-f(x+2*h))/(2*h)
        
    return d

Progressive = ProgressiveDerivative(Function,0.5,h)
print(Progressive)


def CentralDerivative(f,x,h):
    d = 0.
    
    if h != 0:
        d = (f(x+h) - f(x-h))/(2*h)
        
    return d

Central = CentralDerivative(Function,0.5,h)
print(Central)

def ExactDerivative(x):
  return ((1/np.cos(x))**2)/(2*(np.tan(x)**(1/2)))

 
plt.scatter(x,ExactDerivative(x),label='Exact Derivative')
plt.scatter(x,ProgressiveDerivative(Function,x,h),label='Progressive Derivative')
plt.scatter(x,CentralDerivative(Function,x,h),label='Central Derivative')
plt.legend(loc = "upper left")
```

```python
error_progressive = np.zeros(len(x))
error_central = np.zeros(len(x))

for i in range(len(x)):
    error_progressive[i] = np.abs(ProgressiveDerivative(Function, x[i], h) - ExactDerivative(x[i]))
    error_central[i] = np.abs(CentralDerivative(Function, x[i], h) - ExactDerivative(x[i]))
    
    
plt.plot(x,error_central, color='r',label='Error Central')
plt.plot(x,error_progressive, color='y',label='Error Progresiva')
plt.legend(loc = "upper right")
plt.show()


```

```python
import matplotlib.pyplot as plt
```

```python
probabilidades=[]
no_personas=[]
for n in range(0,80):
  probabilidad=1
  for j in range(0,n):
    probabilidad*=(365-n)/365
  probabilidades.append(probabilidad)
  no_personas.append(n)
```

```python
fig, ax = plt.subplots()
ax.plot(no_personas, probabilidades)
```

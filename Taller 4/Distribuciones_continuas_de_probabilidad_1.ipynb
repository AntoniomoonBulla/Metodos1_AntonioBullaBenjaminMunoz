```python
import sympy as sym

x = sym.Symbol('x')
y = sym.Symbol('y')
f = (2/3)*(x+2*y)
```

```python
distr = sym.integrate(sym.integrate(f, (x, 0, 1)), (y, 0, 1))
distr
```

Dado que estaa integral es igual a 1. la funci√≥n es de densidad conjunta

```python
g = sym.integrate(f,(y,0,1))
g
```

```python
h = sym.integrate(f,(x,0,1))
h
```

```python
Ex = sym.integrate(g * x,(x,0,1))
Ex
```

```python
Ey = sym.integrate(h * y,(y,0,1))
Ey
```

```python
Exy = sym.integrate(f*x*y,(x,0,1),(y,0,1))
Cov = Exy - Ex * Ey
Cov
```

```python
Cov_ = sym.integrate((x - Ex) * (y - Ey) * f,(x,0,1),(y,0,1))
Cov_
```

Puesto que la covarianza es diferente de cero, las variables x, y son independientes.

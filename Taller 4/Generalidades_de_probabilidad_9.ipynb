```python
import numpy as np
import matplotlib.pyplot as plt

p1 = np.linspace(0.1,0.9,1000)
p2 = np.linspace(0.1,0.5,1000)
x, y = np.meshgrid(p1,p2)


def Probabilidad(p1, p2):
    return 0.5 * (p1* (1-p2) + p2* (1-p1)) + 0.25 * (p1* p2 + (1-p2) * (1-p1))

P = Probabilidad(x,y)
fig = plt.figure()
ax = fig.add_subplot(projection ='3d')
ax.plot_surface(x,y,P,cmap="plasma")
ax.view_init(elev=20, azim=85)
plt.show()
```

```python
max = 0
min = 1

for i in range(len(P)):
    for j in range(len(P)):
        if P[i][j] >= max:
            max = P[i][j]
        if P[i][j] <= min:
            min = P[i][j]
            
puntomax = []
puntomin = []

for i in range(len(p1)):
    for j in range(len(p2)):
        valor = Probabilidad(p1[i],p2[j])
        if valor == max:
            puntomax = [p1[i],p2[j]]
        if valor == min:
            puntomin = [p1[i],p2[j]]

max = (round(max, 3), puntomax)
min = (round(min, 3), puntomin)
```

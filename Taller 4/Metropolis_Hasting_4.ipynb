```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return 1 / (np.pi * (1 + x**2))

x0 = 0

def metropolis(f,x0, Nsteps= 1000, delta= 0.4):
  
  x = np.zeros(Nsteps)
  x[0] = x0

  for i in range(0, Nsteps-1):
    
    xf = x[i] + np.random.uniform(-delta, delta)

    if np.random.rand() < min(1, f(xf)/f(x[i])):
      x[i+1] = xf
    
    else:
      x[i+1] = x[i]

  return x

MCMC = metropolis(f, x0)

plt.hist(MCMC, density=True, bins=30)
xs =np.linspace(-10,10,100)
plt.plot(xs,f(xs))

plt.show()
```

```python
import numpy as np

x = np.array([0,1,2])

def Combinatoria(r,n):
  return np.math.factorial(n) / (np.math.factorial(np.abs(r))*np.math.factorial(np.abs(n-r)))

def f(x):
  c1 = np.math.factorial(7)/((np.math.factorial(2-x))*np.math.factorial(7-(2-x)))
  c2 = np.math.factorial(3)/((np.math.factorial(x))*np.math.factorial(3-x))
  c3 = np.math.factorial(10)/((np.math.factorial(2))*np.math.factorial(10-2))

  P = c1*c2/c3

  return P

prob = np.zeros(3)

for i in range(len(x)):
  prob[i] = f(x[i])
prob

μ = prob[0]*x[0] + prob[1]*x[1] + prob[2]*x[2]
```

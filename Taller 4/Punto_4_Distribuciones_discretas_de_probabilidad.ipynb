```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion_probabilidad(x,y):

  if (4-x-y)>=0 and (4-x-y)<=3:

    a=np.math.factorial(3)/((np.math.factorial(x))*np.math.factorial(3-x))
    b=np.math.factorial(2)/((np.math.factorial(y))*np.math.factorial(2-y))
    c=np.math.factorial(3)/((np.math.factorial(4-x-y))*np.math.factorial(3-(4-x-y)))
    d=np.math.factorial(8)/((np.math.factorial(4))*np.math.factorial(8-4))

    p=((a*b*c)/d)*100
  else:
      p=0
  return p
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111,projection='3d')

p=np.zeros([3,2])
print(p)
for x in range(0,3):
  for y in range(0,2):
    print(funcion_probabilidad(x,y))
    p[x][y]=funcion_probabilidad(x,y)
    ax.scatter3D(x,y,p[x][y])
```

```python
Exy=0

for i in range(0,3):
  for j in range(0,2):
      Exy+=p[i][j]*i*j
print(Exy)
```

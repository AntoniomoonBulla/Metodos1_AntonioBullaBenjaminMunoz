```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1
def sucesionFib(n):
    suc = np.arange(n)
    i = 1
    j = 0
    for k in range(n):
        b = i+j
        i = j 
        j = b
        suc[k] = j
    return suc
```

```python
x = sucesionFib(20)
x
```

```python
#2


xpoints = np.arange(20)
ypoints = x

plt.plot(xpoints, ypoints)
plt.xlim(0,20)
plt.plot(x, "k", label = "Sucesión Fibonacci")
plt.legend(loc = "upper left")
plt.show()
```

```python
#3

au = (1+np.sqrt(5))/2

from sympy import*
def afib(n):
    i = 1
    j = 0
    suc = np.zeros(n)
    for k in range(n):
        b = i+j
        i = j 
        j = b
        suc[k] = 1+(i/b)
    return suc


plt.plot(afib(20), label = "Estimación usando la serie")
plt.legend(loc = "upper right")
plt.axhline(y=au, color='r',ls='--')
```

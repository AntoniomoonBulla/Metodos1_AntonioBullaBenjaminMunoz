```python
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
x = np.linspace(-4,4,25)
y = np.linspace(-4,4,25)
X,Y = np.meshgrid(x,y)
```

```python
def potflujo(x,y):
    return (2*x*(1-(4/((x**2)+(y**2)))))
```

```python
h=0.001
def dpx(f,x,y,h):
    if h != 0:
        return(f(x+h,y)-f(x-h,y))/(2*h)
def dpy(f,x,y,h):
    if h != 0:
        return(f(x+h,y)-f(x-h,y))/(2*h)
```

```python
Vx= dpx(potflujo, X, Y, h)
Vy=-1*dpy(potflujo, X, Y, h)
a = plt.figure()
ax = a.add_subplot(122)
for i in range(25):
    for j in range(25):
     ax.quiver(x[i],y[j],Vx[i,j],Vy[i,j])
```

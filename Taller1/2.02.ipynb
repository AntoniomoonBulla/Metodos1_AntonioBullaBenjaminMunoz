```python
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/EstrellaEspectro.txt'
```

```python
filename = 'EstrellaEspectro.txt'
```

```python
urllib.request.urlretrieve(url, filename)
```

```python
data = np.loadtxt(filename)
```

```python
plt.plot(data[:,0],data[:,1])
```

```python
puntosx = data[:,0]
puntosy = data[:,1]

#Recorrer eje 'x' y comparar imgenes evaluando iterativamente
#si un punto medio es mayor que su anterior y menor que su inmediato siguiente, es un mximo

maxloc = []
for i in range(1,len(puntosx)-1):
    
    if puntosy[i-1] < puntosy[i] > puntosy[i+1]:
        maxloc.append((puntosx[i],puntosy[i]))
    
maxloc = np.array(maxloc)
plt.plot(puntosx, puntosy)
plt.scatter(maxloc[:,0],maxloc[:,1],marker = '^', color = 'red')
```

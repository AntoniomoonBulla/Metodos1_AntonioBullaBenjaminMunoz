```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

#a

ec = np.array([[2.,-1.],[1.,2.],[1.,1.]])
sol = np.array([2.,1.,4.])

def GetFit(A,b,n=2):
    
    n = A.shape[0]
        
    AT = np.dot(A.T,A)
    bT = np.dot(A.T,b)
    
    xsol = np.linalg.solve(AT,bT)
    
    return xsol

solpoint = GetFit(ec,sol)
x = np.linspace(-5,5,100)

#Se despeja y en las 3 ecuaciones

y1 = 2*x-2
y2 = (1-x)/2
y3 = 4-x


plt.scatter(solpoint[0],solpoint[1])
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
#b

h = 0.01

x = np.arange(-5, 5, h)
y = np.arange(-5, 5, h)


def f(x,y):
  return np.sqrt((2*x-2-y)**2 + (x+2*y-1)**2 + (x+y-4)**2)

f = np.vectorize(f)

x, y = np.meshgrid(x, y)
z = f(x,y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,y,z, cmap='coolwarm')
plt.show()
```
